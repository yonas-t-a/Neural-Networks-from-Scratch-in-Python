```python
class ReLu_Activation:
    def ReLu_forward(self, input):
        return np.maximum(0, inputs)
```

```python
from nnfs.datasets import spiral_data
import nnfs

X, y = spiral_data(samples=100, classes=3)
```

```python
class Layare_Dense(ReLu_Activation):
    def __init__(self, n_input, n_neuron):
        self.weight = 0.1 * np.random.randn(n_input, n_neuron)
        self.bias = np.zeros(n_neuron)
        
    
    def forward(self, input):
        return self.ReLu_forward(np.dot(input, self.weight) + self.bias)
```

```python
myPerceptron =  Layare_Dense(4,1)
inputs = [-1,2,-5,6]
output = myPerceptron.forward(inputs)
output
```
