```python
import numpy as np
```

```python
x = np.array([1,2,3])
w = np.array([0.2, .8, -0.5])
b = 2

output = x[0]*w[0] + x[1]*w[1] + x[2]*w[2] + b

print(output)
```

```python
input = np.array([1,2,3,4])
weight = np.array([0.2, 0.8, -0.5, 1.0])
b = 2


output = input[0]*weight[0] + input[1]*weight[1] + input[2]*weight[2] + input[3]*weight[3]+ b
print(output)
```
